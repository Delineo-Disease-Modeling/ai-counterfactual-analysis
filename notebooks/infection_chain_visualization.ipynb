```python
# to generate infection chain graph based on infection_logs.csv 

import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd 
import math

df = pd.read_csv("../data/raw/sample/infection_logs.csv")
df.head()

G = nx.DiGraph()

for index, row in df.iterrows():
    infector_person = row['infector_person_id']
    if (math.isnan(infector_person) or infector_person == ''): 
        infector_person = 'Initial Seed'
    G.add_edge(row['infected_person_id'], row['infector_person_id'])

plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)  
nx.draw_networkx(G, pos, with_labels=True, node_color='lightblue', node_size=2000,
                 font_size=10, font_weight='bold', arrowsize=20)
plt.title("Infection Chains")
plt.axis('off') 
plt.show()

```
